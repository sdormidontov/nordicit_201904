```python
# настраиваем график, для этого создаем кастомный subplot

fig = plt.figure() # техническая функция, которая достает объект, чтобы сделать subplot. fig - это как холст
axes = fig.add_subplot(111) # переход к нижнему слою. axes - это область в черной рамке, внутри которого рисуем график. 
# Может быть несколько сабплотов

fig.set_facecolor('r')

axes.set(
    facecolor = 'g',
    xlim = [-10, 10],
    ylim = [-5, 5],
    title = 'axis',
    xlabel = 'axis X',
    ylabel = 'axis Y'
)

plt.plot([-1, 0, 1 , 2 , 3 , 4 , 5], axes = axes, color = 'white')
plt.show()
```

```python
years_df = df.groupby(['Year_of_Release'])[['Global_Sales']].sum().join(
    df.groupby(['Year_of_Release'])[['Name']].count()
)
                     
```

```python
years_df.head()
```

```python
years_df.columns = ['Global_Sales', 'Number_of_Games']
```

```python
years_df.head()
```

```python
# создаем линию
trace0 = go.Scatter(
    x=years_df.index,
    y=years_df.Global_Sales,
    name='Global_Sales'
)
```

```python
# создаем линию
trace1 = go.Scatter(
    x=years_df.index,
    y=years_df.Number_of_Games,
    name='Number of games released'
)
```

```python
data = [trace0, trace1]
layout = {'title': 'Statistics of video games'}

fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
plotly.offline.plot(fig, filename='years_stats.html', show_link=False)
```

```python
platform_df = df.groupby(['Platform'])[['Global_Sales']].sum().join(
    df.groupby(['Platform'])[['Name']].count()
)

```

```python
platform_df.columns = ['Global_Sales', 'Number_of_Games']

platform_df.sort_values('Global_Sales', ascending=False, inplace=True)
```

```python
# создаем линию
trace0 = go.Bar(
    x=platform_df.index,
    y=platform_df.Global_Sales,
    name='Global_Sales'
)

# создаем линию
trace1 = go.Bar(
    x=platform_df.index,
    y=platform_df.Number_of_Games,
    name='Number_of_Games'
)

data=[trace0, trace1]
layout = {'title': 'Statistics of platforms', 'xaxis': {'title': 'platform'}, 'yaxis': {'title': 'esdfsdfg'}}

fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```
